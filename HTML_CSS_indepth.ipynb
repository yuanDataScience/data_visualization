#### Semantic and Meta Tags
* Semantic tags
  + semantic tags that describe the meaning of a web page and they help search engines and accessibility softrware like screen readers to understand the contents of a page.
  +  sematic tags are usually inside \<body\> tag, consists of header, main and footer with their internal tag structures
    + header
      + contains company logo and navigation links
      + the main navigation section of your web page can be described by the nav tag
      + \<nav\> element is always after header element and the header element is used for logos
        + navigation links in nav element are specified as ul inside nav
    + main
      + cotains sections and articles
      + the two key semantic elements for main content are the article and section elements
      + article element: a content in a document page application or site that is
        + complete, or self-contained  
        + independently distributable
        + what can be an article? 
          +forum post, article, blog post, user comment, interactive widget, or other independent item of content
        + you can contain a blog post content inside of the article element
          + more specifically, this article tag can contain a h2 and a p tag for header and content
      + main element can contain a set of articles     
    + footer
      + contact information and social media links
      + may contain additional navigation links and contents
* the semantic tags can be nested inside of each other
  + an article element can contain header and main elements
* section element
  + can be added to define individual sections of an article or a webpage
  + must contain heading elements to describe the section
  + you can also use section elements to describe the different sections of your webpage
  + section doesnot require the article elements
* semantic tags in action
  + start from a basic semantic strucure
    + header
    + nav
    + main
    + footer
  + add logo in header, ul in nav, and articles in main (each article is a separate blog), with a h1 heading for Blog

* how meta data helps social media to create preview of other web pages
  + Facebook established the Open Graph Protocol in 2010
  + Open Graph Protocol is a set of metadata rules that allows web pages to describe themselves to social networks
    + social media platform use these meta tags to create a preview of the shared web page
  + OGP uses the property of meta tag to define the metadata name and content to define the metadata value
    + properties must has og: as prefix
  + To set up a social media card, you must alwasy include four og properties
    + og:title
      + defines the title of the page (not the web site) that appears in the preview
    + og:type
      + type of the content, such as website, video, music or article
    + og:url
      + permanent url the social network must use for the specific page (eg. link to the actual post)
    + og:image
      + url of an image that must display when the website is shared
  + other optional og properties are
    + description
    + locale
      + language and country
    + site_name
      + the overall website that the webpage belongs to
     ```html
        <meta property="og:title" content="Our Menu">
        <meta property="og:type" content="website">
        <meta property="og:url"   content="https://www.mysite">        
        <meta property="og:image" content="https://example.com/image.png">

        <meta property="og:description" content="A brief description of Little Lemon's business">
        <meta property="og:locale" content="en_US">
        <meta property="og:site_name" content="Little Lemon">
    ```

* form and validation
  + form validation is the process to ensure that data is valid and conforms to defined rules
  + client-side validation
    + checks errors as soon as they are typed into the form and provide users with imeemdiate feedback
    + done by web browser (javaScript code)
    + HTML has several input types validated by web browser, include
      + email
      + tel
      + url
      + date
      + time
      + number
      + range
      + color
   
  + server-side validation
    + checks errors after the data havd been submitted to the web server
    + more secure as it prevents malicious users from tampering with your website's code and submit invalid date to server
    + can run more complex checks against database or business requirements
  + most web sites use both client and server side validation 
* creating a form
  + to associate a label element to the input elelment, set the for attribute's value the id of the input element
  + after the label element, add a line breaker so that the input element appears below the label
  + organize each label, input element pair inside a div element
  + in the last div element of the form element, add a button with type="submit"
* further client-side validation of form  
  + specify 'required' in an input to prevent users input a blank field for submission
  + specify minlength and maxlength to restrict the minimum and maximum lengths of the input
  + to turn an invalid field to red color, use css psudo class of input:invalid

* Form submission
  + when using GET method, the element's key and values are sent by attaching key/values to the url of the request, and web server will extract these key/value pairs. The problems are
    + URL length is limited by browser which will not allow a lot of parameters to be transferred
    + URL length limited by web server (usually 4096 chars)
    + security threats (parameters transferred may be stored in browser's history)
  + when using POST method the key/value pairs (form content) are embedded in the HTTP request body to be sent to server
    + more secure
    + can use HTTPS to encrypt the data so only the receiver and sender can understand the data
  + the absolute and relative path of action attribute of form element
    + abosolute path will be the base address of the current website combined with the absolute path
      + if the current web page is https://meta.com/company-info, and /login is the absolute path, then the submission address is https://meta.com/login
    + relative path will be the current web address combined with the relative path. If thw web browser is currently on th web page https://meta.com/company-info/, and relative path is login, the submission address will be https://meta.com/company-info/login  
    + when the action attribute is not specified, the form will default to the same URL

#### Flexbox ([resource link](https://css-tricks.com/snippets/css/a-guide-to-flexbox/))
* general rules
  + a container that can grow, shrink and align the items inside it
  + one-dimensional that allows to align along a row or a column (set to row alignment by default)
  + elements inside the container are placed starting from the top left
  + when aligned along the row, the horizonal axis is the main axis and the vertical axis is called the corss axis
  + you can set flex-direction property to flip the main axis and cross axis so that the elements will move down along the vertical main axis (row|row-reverse|column|column-reverse)
  + to use flexbox, set the display property of the flexbox container to flex or inline-flex in css
 
* alignment properties for parent container
  + justify-content: for item alignment on main axis relative to container(default is left)
    + flex-start (default): items are packed toward the start of the flex-direction
    + flex-end: items are packed toward the end of the flex-direction
    + center: items are centered along the line
    + left: items are packed toward left edge of the container, unless that doesn’t make sense with the flex-direction, then it behaves like start.
    + right: items are packed toward right edge of the container, unless that doesn’t make sense with the flex-direction, then it behaves like end.
    + space-between: items are evenly distributed in the line; first item is on the start line, last item on the end line
    + space-around: items are evenly distributed in the line with equal space around them. Note that visually the spaces aren’t equal, since all the items have equal space on both sides. The first item will have one unit of space against the container edge, but two units of space between the next item because that next item has its own spacing that applies.
    + space-evenly: items are distributed so that the spacing between any two items (and the space to the edges) is equal.
  + align-items: for item alignment on cross axis 
    + stretch (default) stretch to fill the container (still respect min-width/max-width)
    + flex-start / start / self-start: items are placed at the start of the cross axis. The difference between these is subtle, and is about respecting the flex-direction rules or the writing-mode rules.
    + flex-end / end / self-end: items are placed at the end of the cross axis. The difference again is subtle and is about respecting flex-direction rules vs. writing-mode rules
    + center: items are centered in the cross-axis
    + baseline: items are aligned such as their baselines align   
  + align-content: for packing flex lines and control space
    + only applies for multi-line flexible containers, where flex-wrap is set to either wrap or wrap-reverse
    + use the same rules applie to justify-content, except for stretch      
  + flex-wrap: wrap items to the size of the available viewport
    + nowrap as default, wrap (top to bottom), and wrap-reverse (bottom to top))
  + flex-direction: set the main axis (column, default is row)
  + gap: create space between items along main axis. can also config column-gap and row-gap
  + flex-flow: shorthand for flex-direction and flex-wrap (flex-flow: column wrap)
    
* properties for the children(flex items)
  + align-self: for unique flex items on cross axis that overrides align-items or default alignment
  + flex: flex-grow flex-shrink flex-basis (flex: 0 1 auto; as default)
  + flex-grow: how much the item will grow compared to other flexible items
  + flex-shrink: how much the item will shrink compared to other flexible items
  + felx-basis: specifies the initial length of the flexible item. (20% or 5rem etc.)
      + auto means look at my width or height property

  + navigation bars

    + highly responsive on different devices
  + image gallery  
    + realign itself as you change the size of the window
    
#### Grid layout
* general rules
  + two dimension
  + responsive and compatible with browser variations
  + suitble for large layout
  + in the following code example
    + column width are defined by pixels
    + the height of rows are defined by fr, which is fraction
      + fraction divide the grid. The first row has a height that is twice of the second row
      + fraction makes it possible to define the relative size without using pixels
      + both pixels and fractions can be used interchangeably for both rows and columns
      + grid-gap is for gutter
      + grid-auto-column and grid-auto-columns: define default sizes for rows and columns if not explicitly specified
      + you can use repeat() function to define 3 100px by repeating 100px 3 times
      + minmax() sets the minimun and max sizes of rows and columns
      + the best way to ensure uniformity of rows and columns when setting grids is to use auto properties and the repeat function together
      
  ```css
    .container {
        display: grid;
        grid-template-columns: 100px 100px 100px; // repeat(3, 100px);
        grid-template-rows: 2fr 1fr; // grid-auto-rows: minmax(150px, auto);
        grid-gap: 10px; 
        background-color: rgb(202, 96, 96);
     }
   ```
* Grid container CSS properties ([link from learncssgrid.com](https://learncssgrid.com/))
  + disply: grid or inline-grid
    + grid: grid items are placed in rows by default and span the full width of the grid container
    + inline-grid: grid items have width decided by content
  + grid-template-rows: 50px 100px
    + set up two row tracks with their heights as 50px and 100px, respectively.
    + if there are more row items, their heights are defined by contents
  + grid-template-colums: 90px 50px 120px
    + set up 3 columns, with widths of 90px, 50px and 120px, respectively
    + if we have more than 3 items, extra items will go to the next row
    + you can also use fr to distribute the space for each column/row
      + fr is calculated based on the remaining space when combined with other length values
      + grid-template-columns: 3rem 25% 1fr 2fr means:
        + 1fr = ((width of grid) - (3rem) - (25% of width of grid)) / 3
      + fr also combines with auto. auto means the size depending on the content size

  + nth-of-type and nth-child selectors

  + star selector selects all the elements on the page (universal selector, has a specificity score of 0)
  + group selectors by grouping elements separated by comma
    + h1, p
* specificity of CSS rules
  + Specificity is the ranking or score that helps CSS determine the final rule that will be applied to a given element
  + there are four categories in the CSS specificity hierarchy
    + inline styles
      + attached to th elements within HTML code. eg. defined as style attribute
    + IDs
    + classes, attributes and pseudo-classes
      + .class
      + p\["attribute"\]
      + div:hover
    + Elements and pseudo-elements
    + example of score calculation
      + \#hello: 0100 (ID)
      + div: 0001 (element)
      + div p.foo: 0012 (two elements and a class)
* combination selectors
  + descendant selectors (using space)
    + elements contained within another selector
    + \#blog h1 applies to all h1 descendants of the element with the id of blog
  + child selectors (using > applies to only direct children)
    +\#blog > h1 select the immediate h1 child of element with the id of blog
  + general sibling selectors (using ~)
    + h1 ~ p: select all p elements having a sibling of h1 element
    + this applies to the pattern where there are p elements following an h1 elements
  + ajacent sibling selectors (using +)
    + only the first or adjacent element that follows another element
    + h1 \+ p select the first p element adjacent to an h1 element
    + p must be the first adjacent sibling of h1
  + class a and b (.a.b)
  + class a in element p (p.a)  
* pseudo-selectors (pseudo-classes)
  + state based selectors
  + in focus
  + visited links
  + specific elements

#### CSS Transforms and transitions
* syntax
```css
element {
    transform: value;
    transition: value;
}
```
* transform:
  + modifies the spatial position of an element
  + change scale and angle
* transition:
  + controls the speed of the effect
  + used in conjunction with transform
* control the rotate transform with a 2s transition
  + this is done by specifying the transition of transforms to be 2s
```css
h1 {
    transform: rotate(0deg);
}

h1:hover {
    transform: rotate(20deg);
    transition: 2s;
}
```

#### CSS animation
* combine animation property with @Keyframe rule to create advanced animations
* use animation property and control the steps of animation by @Keyframe rule
* syntax
  + you can use the optional from and to to define the steps of animations
```css
@keyframes animationname {keyframes-selector {css-styles;}}

@keyframes animationname {
    from { color: blue; }
    to {color: red; }
}
```
* animation name is an animation property. Animation has the following properties:
  + name
  + duration
  + timing-function
  + delay
  + iteration-count
  + direction
  + fill-mode
  + play-state
* you can use the shorthand of animation to define the animation properties, and leave others as default
```css
animation:
div {
    animation: myanimation 3s infinite; /* name duration timing-function */
}
```
* code example of minute and hour clock hands
```css
.minutes {
    animation: cycle 6s infinite linear;
}
.hours {
    animation: cycle 60s infinite linear;
}

@keyframes cycle {
    from { transform: rotate(0deg); }
    to { transform: rotate(360deg); }
}
```
* keyframes
  + using @keyframes
  + can use 'from to' to define the start and end of the transformation
  + for more detailed controls, can use % to define the intermediate states
  + can apply to different properties during the transition
  + tied to animation-name, which is required, as well as animation-duration. Other properties can be added on
  ```css
          body{ 

        padding: 30px;}  

        .box{ 

        background-color: lightcoral; 

        width: 50px; 

        height: 50px; 

        animation: myanimation 3s infinite ease-in; 

        } 

        @keyframes myanimation{
        0%{width: 50px;
        }

        50%{background-color: aquamarine;
            height: 20px;
        }

        100%{width: 100px;
        }

    ```
* code example of a bouncing ball
```css
@keyframes bounce {
    from {
        transform: translate3d(0, 0, 0);
    }
    
    to {
        transform: translate3d(640px, 0, 0);
    }
}

#ball {
    width: 80px;
    height: 80px;
    border-radius: 50%;
    background-color: #62687F;
    animation: bounce 2s linear;
    animation-direction: alternate;
    animation-iteration-count: infinite;
}
```
