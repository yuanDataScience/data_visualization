#### Semantic and Meta Tags
* Semantic tags
  + semantic tags that describe the meaning of a web page and they help search engines and accessibility softrware like screen readers to understand the contents of a page.
  +  sematic tags are usually inside \<body\> tag, consists of header, main and footer with their internal tag structures
    + header
      + contains company logo and navigation links
      + the main navigation section of your web page can be described by the nav tag
      + \<nav\> element is always after header element and the header element is used for logos
        + navigation links in nav element are specified as ul inside nav
    + main
      + cotains sections and articles
      + the two key semantic elements for main content are the article and section elements
      + article element: a content in a document page application or site that is
        + complete, or self-contained  
        + independently distributable
        + what can be an article? 
          +forum post, article, blog post, user comment, interactive widget, or other independent item of content
        + you can contain a blog post content inside of the article element
          + more specifically, this article tag can contain a h2 and a p tag for header and content
      + main element can contain a set of articles     
    + footer
      + contact information and social media links
      + may contain additional navigation links and contents
* the semantic tags can be nested inside of each other
  + an article element can contain header and main elements
* section element
  + can be added to define individual sections of an article or a webpage
  + must contain heading elements to describe the section
  + you can also use section elements to describe the different sections of your webpage
  + section doesnot require the article elements
* semantic tags in action
  + start from a basic semantic strucure
    + header
    + nav
    + main
    + footer
  + add logo in header, ul in nav, and articles in main (each article is a separate blog), with a h1 heading for Blog
  ```html
    <body>
        <header>
            <img src="logo.png">
        </header>
        <nav>
            <ul>
                <li><a href="index.html">Home</a></li>
                <li><a href="location.html">Location</a></li>
                <li><a href="blog.html">Blog</a></li>
            </ul>
        </nav>
        <main>
            <h1> Blog </h1>
            <article>
                <h2>20% off this weekend</h2>
                <p>Come down to Little Lemon and your meal will be 20% off</p>            
            </article>
            
            <article>
                <h2>Our new menu</h2>
                <p>We've updated our menu with new dinners.</p>            
            </article>
        </main>
          <footer>
              <p>Copyright Little Lemon</p>            
           </footer>  
  </body>  
           
```           

* metadata
  + search engines
    + search engine optimization (SEO) is a process to make improvements to a website's content, semantics and delivery to improve its ranking in search results
    + search engine analyzes the html document, media content and links of your website, and rank the website for various seach terms
    + every searc engine has its own algorithm for ranking websites
  + meta tags define metadata about a web page
    + meta tag is inside the head tag and is not visible to the browser.
    + there is no closing tag in meta tag
    + meta tag has two attributes, name and content, representing the name and value of the meta data, respectively
      + examples include:
        + name ="author" content="jane WIlson"
        + name ="description" content="Jane's first web page"
        + name = "robots" content="index,follow"
      + robots metadata tells the search engine if and how to analyze the web page. content can have 4 values 
        + index: tell the bot to analyze the page
        + follow: also visit all links on the web page
        + noindex: tells bot not to analyze the page
        + nofollow: not to visit the links on the page
        
    ```html
        <head>
            <meta name="" content="">
        </head>
    ```
      + viewport metadata (in meta tag with name="viewport", and content="width=device-width, initial-scale=1.0")
        + important when designing responsive web pages  
  + also has meta http-equiv, and meta charset

* how meta data helps social media to create preview of other web pages
  + Facebook established the Open Graph Protocol in 2010
  + Open Graph Protocol is a set of metadata rules that allows web pages to describe themselves to social networks
    + social media platform use these meta tags to create a preview of the shared web page
  + OGP uses the property of meta tag to define the metadata name and content to define the metadata value
    + properties must has og: as prefix
  + To set up a social media card, you must alwasy include four og properties
    + og:title
      + defines the title of the page (not the web site) that appears in the preview
    + og:type
      + type of the content, such as website, video, music or article
    + og:url
      + permanent url the social network must use for the specific page (eg. link to the actual post)
    + og:image
      + url of an image that must display when the website is shared
  + other optional og properties are
    + description
    + locale
      + language and country
    + site_name
      + the overall website that the webpage belongs to
     ```html
        <meta property="og:title" content="Our Menu">
        <meta property="og:type" content="website">
        <meta property="og:url"   content="https://www.mysite">        
        <meta property="og:image" content="https://example.com/image.png">

        <meta property="og:description" content="A brief description of Little Lemon's business">
        <meta property="og:locale" content="en_US">
        <meta property="og:site_name" content="Little Lemon">
    ```

#### User Input and Forms

* form and validation
  + form validation is the process to ensure that data is valid and conforms to defined rules
  + client-side validation
    + checks errors as soon as they are typed into the form and provide users with imeemdiate feedback
    + done by web browser (javaScript code)
    + HTML has several input types validated by web browser, include
      + email
      + tel
      + url
      + date
      + time
      + number
      + range
      + color
   
  + server-side validation
    + checks errors after the data havd been submitted to the web server
    + more secure as it prevents malicious users from tampering with your website's code and submit invalid date to server
    + can run more complex checks against database or business requirements
  + most web sites use both client and server side validation 
* creating a form
  + to associate a label element to the input elelment, set the for attribute's value the id of the input element
  + after the label element, add a line breaker so that the input element appears below the label
  + organize each label, input element pair inside a div element
  + in the last div element of the form element, add a button with type="submit"
* further client-side validation of form  
  + specify 'required' in an input to prevent users input a blank field for submission
  + specify minlength and maxlength to restrict the minimum and maximum lengths of the input
  + to turn an invalid field to red color, use css psudo class of input:invalid
  
  ```html
    <form>
        <input type="text" id="user_name" name="user_name" required minlength="3" maxlength="12">
    </form>
    
    <style>
        input:focus:invalid {
            border: 2px solid red;
        }
    </style>
   ```

* Form submission
  + when using GET method, the element's key and values are sent by attaching key/values to the url of the request, and web server will extract these key/value pairs. The problems are
    + URL length is limited by browser which will not allow a lot of parameters to be transferred
    + URL length limited by web server (usually 4096 chars)
    + security threats (parameters transferred may be stored in browser's history)
  + when using POST method the key/value pairs (form content) are embedded in the HTTP request body to be sent to server
    + more secure
    + can use HTTPS to encrypt the data so only the receiver and sender can understand the data
  + the absolute and relative path of action attribute of form element
    + abosolute path will be the base address of the current website combined with the absolute path
      + if the current web page is https://meta.com/company-info, and /login is the absolute path, then the submission address is https://meta.com/login
    + relative path will be the current web address combined with the relative path. If thw web browser is currently on th web page https://meta.com/company-info/, and relative path is login, the submission address will be https://meta.com/company-info/login  
    + when the action attribute is not specified, the form will default to the same URL

* iframe
  + allows you to place or embed content from another website into a web page
  + it embeds another broswering instance inside the page, similar to run another webpage like running it in another tab
    + meaning that it runs HTML, CSS and JavaScript inside that webpage
  + use allow attribute to disable camera and microphone access in the iframe
  + use sandbox attribute limits behavior within the iframe such as prevent files from downloaded and pop-up windows
    + when the attribute is set to "", all sandbox restrictions are applied to the iframe
    + you can allow the behavior by setting sandbox="allow-downloads"
    
  ```html
    <iframe src="https://www.example.com" width="200" height="150" sandbox
            allow="camera 'none'; payment 'none'; microphone 'none';">
    </iframe>
  ```
  + not recommended due to security concerns and need to ensure you trust the website that you are embedding

### Creating layouts
#### CSS web layout
* CSS can modify the style format of 
  + Fonts
  + color
  + Layout
  + Size
* important CSS elements  
  + viewport
    + the browser window that is visible to the user on the screen is called viewport
  + display property
    + specifies the type of box that you want to use for a given HTML element
    + baiscally, you can use box, flex and grid as your layout options
  + flexbox 
    + short for flex box mode
    + + adds responsiveness to CSS with float elements and positioning
    + one dimensional, meaning that a given flex box container will arrange items in either a column or a row along its axis
    + flex box container applied over an element can felx to shrink or expand, resulting in a flexible responsive design
    + more suitable to create flexible elements in smaller spaces
    
    ```css
        #sample {
            display: flex;
    }
    ```
  + grid
    + similar to flexbox, but it creates a two dimensional grid along both the row and column axes
    + can lead to greater complexity later, if the element rules are not systematically defined
    + more suitable to large scale layouts
    ```css
        #sample {
            display: grid;
    }
    ```

* Flexbox
  + a container that can grow, shrink and align the items inside it
  + one-dimensional that allows to align along a row or a column (set to row alignment by default)
  + elements inside the container are placed starting from the top left
  + when aligned along the row, the horizonal axis is the main axis and the vertical axis is called the corss axis
  + elements will move along the main axis and continue to the next row/column when the main axis is filled
  + you can set flex-direction property to flip teh main axis and cross axis so that the elements will move down along the vertical main axis
  + to use flexbox, set the display property of the flexbox container to flex in css
  + alignment properties
    + justify-content: for item alignment on main axis (default is left)
    + align-items: for item alignment on cross axis (flex-end)
    + align-self: for unique flex items on cross axis
    + align-content: for packing flex lines and control space
  + flex-wrap: wrap items to the size of the available viewport
  + flex-direction: set the main axis (column, default is row)
  + gap: create space between items along main axis. can also config column-gap and row-gap
  + flex: flex-grow flex-shrink flex-basis (flex: 0 1 auto;)
    + flex-grow, how to expand when extra free space is available
    + flex-shrink
    + felx-basis: sets the initial size
