### Functional programming

#### Introduction to functional programming
* There are two programming paradigms as different approaches to writing code
  + functional programming
    + clear distinction between data and functions, and data exist outside of functions
    + you pass data to functions in form of arguments, which may return data as the result
    + the returned value of one function can be used as the argument passed to another function
    + in javascript, functions are first-class citizen, which means a function is a value that we can:
      + pass to other functions
      + save in a variable
      + return from other functions
        + the following code shows how we can pass functions by their names to variables, and invoke them using variables
        
        ```javascript
        function addTwoNums(a, b) {
            console.log(a + b)
        }
        function randomNum() {
            return Math.floor((Math.random() * 10) + 1);
        }
        function specificNum() { return 42 };

        // create getNumber variable to save functions
        var useRandom = true;

        var getNumber;

        if(useRandom) {
            getNumber = randomNum
        } else {
            getNumber = specificNum
        }

        addTwoNums(getNumber(), getNumber())
        ```
    + high order functions are functions that have either one or both of the following characteristics:
      + it accepts other functions as arguments
      + it returns functions when invoked
      + the following is a code example of a high order function accepting getNumber as function arguments
      ```javascript
        function addTwoNums(getNumber1, getNumber2) {
            console.log(getNumber1() + getNumber2());
        }
      ```
    + pure function and side-effects
      + A pure function returns the exact same result as long as it's given the same values
      + another rule for a pure function is that it should not have side-effects, which is any instance where a unction makes a change outside of itself. This includes:
        + changing variable values outside of the function iteself, or even relying on outside variables
        + calling a Browser API (even the console itself)
        + calling Math.random() since the value cannot be reliably repeated
      
  + object oriented programming
    + combine data and functions into objects
    + instead of generating new return values by function, methods update properties stored in the object
    + you can make as many objects as you want
* scope in javascript
  + variables defined outside of a function is call global scope
  + variables defined in the function is called local variable, and are scoped to the function
  + use let and const to define block scope variables
    + these variables only exist inside the scope of the curly braces where they are defined
    + const variable must be assigned a value during declaration
    + unlike var, let and const can not do the following
      + can not use them before decalring them (in var, you can put var statement after the statement using the variable)
        + ReferenceError 
      + can not re-declare the variable (syntax error)
      + can not be accessed beyond the block

* prototype
  + a prototype is an object that can properties to be shared by multiple other objects
    + this is the basis of how inheritance works in JavaScript
    + Javascript implements a prototype of inheritance model
  + how the prototype is implemented?
    + by using Object.create(parent_object), all the properties of parent_object are shared by other child objects
      + all child objects have the parent_object as their prototype
    + in the following code example, we create penguin1 using bird as the prototype, and then set its canFly property to False
      + Javascript searchs the property from the object itself, if can not find, will find it from its prototype.
      + the canFly property is in the bird prototype of penguin1 object
      + note that once a property is found along the chain, it will not do further search
      + the new value set to canFly property only affects the penguin1 object, not bird or eagle1 object
      
    ```javascript
    
    var bird = {
        hasWings: true,
        canFly: true,
        hasFeather: true
    }
    
    var eagle1 = Object.create(bird);
    console.log(eagle1.hasWing);

    var penguin1 = Object.create(bird);
    penguin1.canFly = false;
   ```
    
* although it is possible to build inheritance using Object.create() method, it is better to use class syntax for more complex objects and inheritance
  + under the hood, this syntax still works with prototypes
  + after you define a class, you use keyword 'new' to instantiate class object instances
  + in class syntax, 
    + the prototype only contains constructor, the new methods and methods overriden in the current class
    + the self contains all the properties initialized in constructor
    + all the other methods inherited from the base class are in prototype's dunder prototype property
  
* you can define static properties and methods, but these static members can only be accessed by the class. Instance objects do not have the access to these members
* you can define methods in construct function, assign that method to a variable and still have the access to 'this' in non-strict mode using prototype. Here the two methods 
  + note tha all methods defined in a class are in strict mode
```javascript
    // describe the non-strict mode to define instance and static methods 
    // that are bounded with 'this' pointer using function constructor
    function Animal() {}

    //this defines speak as a prototype function for all Animal instance objects
    Animal.prototype.speak = function () {
      return this;
    }

    // this defines static method of the Animal class
    Animal.eat = function () {
      return this;
    }

    const obj = new Animal();
    const speak = obj.speak;
    speak(); // global object (in non–strict mode)

    const eat = Animal.eat;
    eat(); // global object (in non-strict mode)
```

#### JSON string and regular Javascript object
* convert JSON to JS object
  ```javascript
      const jsonStr = '{"greeting": "hello"}'
      const aPlainObj = JSON.parse(jsonStr); // {greeting: 'hello'}
      aPlainObj.greeting = 'hi'
      aPlainObj // {greeting: 'hi'}
  ```   
  
* convert JS object to JSON
    ```javascript
        const data = {
            firstName = "John",
            lastName = "Doe",
            greeting: "Hello"
        }
        JSON.stringify(data)
    ```
* javascript objects can contain functions, JSON strings can not
  + when stringify a javascript object to a JSON, the methods of the object are excluded

`````python
# nonlocal variable used in net functions
def outer():
    x = 5
    
    def change_x():
        nonlocal x
        if x == 0:
            return
        print(x)
        x -= 1        
        change_x()
    change_x()    
outer()        
`````
