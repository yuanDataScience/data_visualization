#### UX
* UX is user experience
  + includes everything related to all the interactions that a user has with a company, its services and products, such as
    + how does something like a remote control, a phone, an app or a website feel to use?
    + how do you feel before, during and after interacting with the product?
    + can user achieve what they want easily from a product?
    + Does the interface provide a smooth, enjoyable experience, making your users feel in control?
    + why do customers keep coming back for more?
* process
  + UX has been defined to a process
  + the process is iterative
    + you may have to go back to previous steps to adjust your design to suit the customers and business goals several times
    + a typical sequence of taskes in the process including:
      + observe users perform a task
      + identify problems from the observation
      + generate ideas to solve the problems
      + prototype your solution
      + test solution
    + repeat each of these stages and steps withing them and test your design again and again until you have something that is usable and enjoyable for everyone

* usability
  + an essetial concept in UX and UI design
  + how intuitive or easier product is to use
  + five usability components for evaluating usability (Nielsen's usability quality components)
    + learnability
      + when customers are trying to order for delivery, the process should be easy to learn from the first time they do it 
    + efficiency
      + when users want to change orders, can they do it quickly and efficiently?
    + memorability
      + what is the user gets distracted?
        + Is it easy to remember where they were when they return?
        + how quickly can they find where they were?
    + errors
      + what if a user makes a mistake? The design should provide solutions to these mistakes and address them before they happen
    + satisfaction
      + is the website pleasant or satisfying to use?
      + Do users enjoy using it?
      + is it easy to use?

* UI and UX artifacts
  + Artifacts are documents that design teams create to gather, organize and present information throughout the UX process, including:
    + User personas
      + a model or character that portrays a possible user of your website or app
      + Personas aid the design team in focusing on the end users while designing a product
      + Personas help a product team constantly focus on their target users, ensuring that the designed product fits their needs and requirements
      + developing user personas helps determine the following users' characters
        + range of users
        + age group
        + sociometric positions
      + user research methods to create user persona:
        + conducting customer interviews
          + interview at least five people
          + stop when you are getting very little or no new information
          + make use of information you already have (you may have them from previous research or know something for sure) 
        + making assumptions
          + if you have experience to make assumptions about customers. Make sure to back them up later with research
        + using web analytics
          + Web analytics tools are an excellent resource for quantitative data
          + They can tell you how your customers behave but not why, and you'll still need to talk to them to find out what motivates them 
    + user stories
      + A user story is a brief statement or abstract that describes the user and their need or goal
      + It establishes who the user is, what they require, and why they require it
      + Each user persona typically has one user story, which usually adhere to a strightforward template:
        + As a (type of user), I want (a goal) for (some reason)
          + example:  "As an online banking customer, I want to add a payee to my account, so I can transfer money to the payee."
      + User stories assist in documenting helpful information about users, such as the various needs and motivations for using a website or app
      + They also help the development team estimate the timeline required to deliver the final product
    + scenarios
      + A scenario is a situation that depicts how users interact with your website or app
      + Scenarios describe the user's motivations for being there (their task or goal) and a question they need to be answered
      + They also suggest possible ways to achieve these goals
      + Scenarios are essentially an extension of the user story and can be applied to various target users
      + They can also be divided into use cases, which describe the sequence of tasks that any given user performs in a provided functionality or path
        + For example, a scenario could describe how the user transfers money to a payee using a mobile device on his way to work
      + Scenarios help stakeholders visualize the design team's ideas by providing context for the intended user experience
      + process
        + Scenario mapping is the first step in scenario planning
          + The design team, developers, and product owners gather to design a plan centered on their user personas
          + They think about the critical task that the user hopes to achieve after defining the primary user through persona development
        + The following step is to conduct a scenario analysis, contextualize the user's goals and walk through the steps the user would take   
    + storyboards
      + A storyboard is a visual representation of how a user will interact with a product
      + Designers can create different storyboards: sketches, illustrations and screenshots, slideshows and animations, or live demos
      + Visualizing a design idea with an interactive storyboard, like high-fidelity prototypes closely resembling the final product, will help the audience remember, empathize, and connect with it
      + process to create: 
        + set the scene by defining:
          + your persona
          + the environment (where the persona is)
          + the plot (what they want to achieve)
        + sketch out the basic idea for each scene and build it up with as many interactios as you want  
* Overview of UX process
  + UX is a process driven discipline with different models
  + following key steps will help with rich user experience design (these steps are iteractive)
    + empathize
      + interview users whith their experience of using the website and listen to their frustrations
        + the purpose is to understand the users' needs and create a persona that you can reference throughout the design process
      + also create an empathy map, which is a scenario and journey map to empathize more with this persona
        + the purpose is to keep your ideas grounded to avoid making assumptions
    + define
      + collect and distill all the information from users and identify the key problems and needs that they are having
      + prioritize these frustrations and pain points by importance
      + by doing these, you now know
        + who the users are
        + what are their frustrations
        + solutions to what problems you need to solve 
      + you then create a use needs statement that clearly outlines the needs your users have for then ideate step  
    + ideate
      + generating ideas by (also iterative process)
        + sketching, brainstorming, mind mapping and handwriting notes
        + keep an open mind and not commit to one specific idea 
      + develop ideas into wireframes
        + two dimensional representation of the user interface which is concerned with 
          + the placement and hierarchy of content and functionalities that are offered, 
          + expected actions that users will take   
    + prototype
      + simulation of the final product
      + starting from wireframes, and give it some color, add buttons and text, and make it interactive
      + simulate scenario to help customers to achieve their goals
    + test
      + present your solution to users and get their feedback
      + customers interact with prototype and try to achieve the tasks
      + any frustrations can be communicated and highlighted at this stage
    + build
      + you start to build since
        + you have listened to users, empathized with them and aim to solve their needs through iterative design techniques
        + you watched them using the product, tweaked product to make it easier and more intuitive

* customer journey map
  + present steps to complete a task and identify emotions as you track the UX goals

* custom journey maps usually consist of the following 5 elements
  + actor
    + The persona or user who goes on the journey
    + You create an individual map for each actor you have
  + scenario and expectations
    + The scenario describes the situation the journey map addresses and is associated with an actor's goal or need and specific expectations
    + For example, Tilly is ordering a meal online. And her expectation includes being able to easily find all the information required to make a decision and complete her order
  + journey phases   
    + The journey phases are the various high-level stages of the journey. Each phase is a user's step when trying a task
    + each pahse is organized by actions, thoughts, and emotions
  + actions, attitudes and emotions
    + actor's behaviors, thoughts, and feelings throughout the journey, mapped within each journey phase
    + a narrative of the actor's actions during that phase  
    + At various stages, mindsets correspond to users' thoughts, questions, motivations, and the information they need. Ideally, these are direct quotes from customers    
  + opportunities
    + Opportunities, along with additional contexts such as ownership and metrics, are mapping insights that inform how the user experience can be improved
    + Insights and opportunities assist the team in extracting knowledge from the map

### Best practice principles
#### Form design
* clarify each questions to make sure users understand the questions
* use clear layout
  + space user input fields consistently so that users can see each field is unique, and identify the labels of inputs
  + create logical grouping to make the form more understandable
  + payment is separated from the address and each row is given a subtle contrast color to differentiate
* allow users to go back 
* visibility of system status and how long it will take to complete the form
  + add indicators of which part the user is currently working on and the percentage of completion in the flow
* add form validation during flow and validation as it happens
  + users can see if the current input is valid or not
* use clear and specific language
* provide guidance for users to compile passwords (valid password rules)
  + provide strength indicator of the password to users with immediate feedback on their choice

### Figma fundamentals
#### Frames, layers and basic shapes
* Frame
  + also referred to as an Artboard where your design lives
  + think of it as a container where your design lives
* layer
  + an object within a frame
  + as you add an object to a frame, a new layer will be created
* group
  + multiple objects are combined in a group
  + once objects are combined in a group, a single top layer is created, which contains all the objects
* frames and groups
  + very similar
  + helps you organize your files by nesting layers or grouping layers together
  + frame provides a lot more functionality like independent sizing
  
#### Figma practice
* create a frame (either directly draw after clcking frame icon, or select a frame type from right-hand side after cliing fram icon)
* draw shapes inside a frame
  + click the shape drop down (the icon is like a circle)
  + select the shape
  + for square and circle, hold the shift key when drawing by rectangle or elipse 
  + duplicate a shape, select the shape and press (ctr or cmd) D on the keyboard 
* right hand panel
  + you can get the coordination, width, height and corner radius of the shape from the righ hand panel
  + you can edit fill, stroke and effects of the shape
* left hand panel to control layers
  + each object/shape has its own layer
  + you can control the overlapping of shapes by changing the order of layers
  + you can lock a layer to lock the shape so that you can not move it
  + you can select several shapes together and group them by right clicking and click 'group selected'
    + a new layer will replace all the shapes' layers. This new layer contains all the shape objects
    + you still can modify the shape in the grouped layer, by double click the shape inside the layer

* Typography is the layout and arrangement of content using pre-made letter systems such as typefaces and fonts
* typography terminology
  + typeface: also known as font family. Made up of fonts
  + a typeface has multiple font weights and a consistent style across all characters, numbers, and symbols
  + typefaces include Arial, Times New Roman and Comic Sans
* typography elements
  + baseline
    + The baseline is the point at which the text line rests
  + cap height
    + The cap height of a typeface refers to the height of the capital letter of flat letters such as M, H, T, or I
    + cap height is less than the typeface's maximum height
  + X-height
    + also known as corpus size, is the distance between the lowercase "x baseline" and meanline
  + ascenders and descenders
    + An Ascender is a part of a letter that extends above the x-height or cap height, such as the stem of a lowercase b or d
    + A descenders are vertical downward strokes that extend beyond the baseline, such as the lowercase g or p
  + weight
    + The overall thickness of a typeface's stroke is referred to as its weight
  + tracking
    + Tracking is the spacing between all a font's characters
  + kerning
    + Kerning is the space that exists between two specific letters
  + leading
    + The space between two lines of text is known as leading (line height)
      + sum of both font height and the sapce 
      + distanct between the bottoms of two lines
  + white space
    + the space between blocks of text
  + stroke
    + A straight or curved line forms the main body of a letter
  + Serif
    + A stroke or foot-like element is connected to the end of the main strokes of some typefaces
  + Sans serif
    + A typeface that does not have any strokes or extra elements at the bottom of a letter

  + use a simple formula based on font size: multiply it by 30

* avoid text walls
  + A wall of text, also known as a gray wall effect, is a single block of text that is devoid of any white space or other design elements
  + By inserting paragraph breaks at appropriate intervals, you can avoid the gray wall effect
  + You can also use the following design elements: images, quotations, headings and subheadings and bullet points
  + When working with long paragraphs, try to include anything that isn't text and breaks up the solid block

#### Usability testing
* what is a usability testing
  + allows key stakeholders and the project team to understand better how people interact with a product
  + use actual people to track actions and reactions (doesnot use personas)
  + crucial to ensure you create an effective, efficient and delightful experience for your users
  + Users are given tasks to perform while being observed by a research to see where they run into difficulties or become confused
  + If many people run into the same issues, recommendations can be made to fix these usability issues, which offers opportunities to make improvements
* components of a usability testing
  + facilitator
    + gives tasks, monitors behavior, listens to feedback, asks follow-up questions
  + participant
    + a user of the product you have desined or a similar product
    + encouraged to think aloud while completing tasks, allowing the facilitator t follow their progress, knowing which tasks the user is completing and the participant has read the instructions correctly
  + task
    + practical actions that the participant might carry out in everyday life, that can be either detailed or vague
    + task phrasing is crucial
      + Minor inaccuracies in the tasks wording can lead participants to misinterpret what they are required to complete or affect how they carry out the task
    + the facilitator may read the task instructions to the participants or they may be given to the participants on writing in task sheet (read tasks or written tasks)
* types of usability testing
  + qualitative testing
    + aims to gather insights, results and narratives about how users interact with the product
    + the most effective method for identifying issues with the user expereience
    + compared to quantitative usability testing, this type of testing is more prevalent
    + the Neilson Norman group advises utilizing five participants for typical qualitative usability research of a single user group to identify the volume of a product issues
  + quantitative testing
    + aims to gather metrics that capture the user experience
    + tasks success and time on task are two metrics frequently collected
    + the most effective method for gathering benchmarks
  + Online usability testing
    + popular because in-person research usually requires more time and money
    + you can have moderated and unmoderated online usability testing
    + operates in a similar way as in-person usability testing
      + In an in-person environment, facilitator assigns the task speaking with the participant
      + Online facilitated tests can typically be held using screen-sharing applications
      + remote unfacilitated usability tests donot have the same participant facilitator interaction
        + facilitator uses online remote testing technology to deliver the tasks
        + participant works alone and at their own pace to do the tasks
        + a session recording and metrics like task success are sent to the researcher when the participant completes the test

* micro-interactions and micro-animations
  + Micro-interactions begin with a trigger and progress to a set of instructions represented visually through animations
    + The visual output of the micro-interaction is the animation
  +  benefits of micro-interactions
    + Provide feedback on the outcome of an action
    + Complete a small task
    + Increase the feeling of direct manipulation
    + Assist users in seeing the outcome of their actions
    + Avoiding user errors
  + four key components when designing mocro-interactions
    + Triggers
      + manual triggers by users
      + system triggers when the system meets a condition automatcially
    + Rules
      + instruct the app on what to do after a trigger event occurs
    + Feedback
      + notifies the users of the trigger's response. Two factors influence feedback:
        + what information does the user require?
        + how often does it have to happen?
    + Modes and Loops
      + when there is a demand, modes are used
      + loops determine how long the micro-interaction must continue and whether it needs to be repeated
  + micro-animations
    + improve intraction and clarify interfaces
    + simple user intefaces
      + Because the user can see something directly happening when using micro-animations, the user gets a sense of direct manipulation
    + context and change
      + It is challenging for a developer to fit all the information on a small screen. Therefore, some elements are sometimes hidden. Micro-animations can assist users in navigating and viewing all the information on a screen
    + user concentration
      + A user can quickly become distracted by moving objects on the screen. As a result, animating a button will prompt the user to click it to start an action on the page
      + Using too much animation flattens the hierarchy and weakens the message that you may want to convey to the use
    + rewards
      + Micro-animations are very useful for rewarding the user after completing an action, such as submitting a form
