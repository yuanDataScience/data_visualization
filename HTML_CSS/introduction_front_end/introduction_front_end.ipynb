### Core internet technologies
* IP (Internet Protocol) defines IPV4 and IPV6 IP address
  + IPV4 contains 4 octets separated by peroids or dots
  + IPV6 contains 8 hexadecimal digits separated by :
  + the data are sent a in series of messages called IP packets
    + IP packets contains IP header and IP data (payload)
    + IP header contains both destination's and source's IP addresses
  + IP packets may have the following three issues:
    + arrive out of order
    + become damaged or corrupted
    + dropped or lost during transit
  + to solve the problems of IP packet transition, payload contains other protocols
    + these protocols are contained inside the IP packet. The two major protocols are
      + Transmission contro protocol (TCP)
        + can resolve all the three issues of data packet tranmission at the cost of a small delay
        + used for sending data that must be correct and in correct order, such as text and images
      + User datagram protocol (UDP)
        + solve the damaged or corrupted packet issue, but packets can still arrive out of order or not arrive at all
        + used for sending data that can tolerate data packet loss, such as voice or live video streaming
* HTTP protocol is used to communicate with web server  
  + HyperText Transfer Protocol
  + used to transfer web resources such as HTML documents(web pages), images and other files
    + HTML documents are sent from webserver to client
    + images and files are transferred to and from the web server using HTTP
  + is a request-response base protocol
    + clients (eg. a browser) send HTTP requests to server
    + server sends HTTP response to clients
  + HTTP requests consist of method path version, headers and body
    ``` text
    GET / HTTP/1.1
    Host: developer.mozilla.org
    Accept-Language: en
    ```
      + method describes the type of action that the client is to perform
        + GET: retrieve information from server
        + PUT: update records on the server
        + POST: send data to server
        + DELETE: remove the resource from server
      + path:
        + representation of where the resource is store on the webserver
      + version of HTTP protocol
        + 1.1 and 2.0 are the most commonly used versions    
      + headers
        + extra information about the request and the client making the request
      + body of conetent the client is sending
    + HTTP Response
      + contains similar method, version and headers
      + message body consisting of HTML document, images and files
      + Status code (in the header) as an integer from 100-599
        + informational: 100-199. these are provisional, interim response. eg. 100 CONTINUE
        + successful: 200-299.request successfully processed. eg. 200 OK 
        + redirection: 300-399. resoruces have been moved. eg. 301 moved permanently
        + client error: 400-499.
          + 400: client sents bad data to server
          + 401: user must log into an account before the request can be processed
          + 403: request is valid, but webserver refused to process it. Usually indicates that the client does not have the sufficient permission to execute an action in a web application
          + 404: request resource was not found on the webserver 
        + server error: 500-599.
      + Staus message, text message to interpret status code
* HTTPS
  + secure version of HTTP
  + secure communications between two computers by encryption
  + encrypt the data before sending and only receivers that can decrpyt the code can see the data

Box Model
* browser allocate a rectangle/box to each element and css rules are applied to the boxes of the elements
* each box contains 4 parts: content, padding, border and margin
  + content: actual content of the element (text or image) with its size defined as content width and content height.
    + browser will calculate the height and width based on the content itself
    + you can use css rules:Content size to define
      + width: 1px;
      + min-width: 1px;
      + max-width: 2px;
      + height: 1px;
      + min-height: 1px;
      + max-height: 2px;      
  + padding: extends content size. Its sizes are padding box width and padding box height
    + tickness of the padding id defined by 
      + padding-top: 4px;
      + padding-bottom: 1px;
      + padding-left: 5px;
      + padding-right: 2px
    + padding box width = content width + padding left-side + padding right-side
    + padding box height = content height + padding top + padding bottom
  + border: goes around padding and content. its sizes are border box width and border box height
    + different styles of borders
      + solid border
      + dashed border
      + dotted border
      + double border
    + border box width = padding width + border left-side + border right-side  
    + border box height = padding height + border top-side + border bottom-side
    ```css
 h1 {
     border-width: medium;
}
```
  + margin extends border area to separate elements. Its sizes are margin box width and margin box height
    + define the size of margin by
      + margin-top: 4px;
      + margin-bottom: 1px;
      + margin-left: 5px;
      + margin-right: 2px
    + margin box width = border width + margin left-side + margin right-side  
    + margin box height = border height + margin top-side + margin bottom-side  
  + using shorthand properties to define padding, border and margin
    + Padding: 4px 2px 1px 5px;
    + Border-width:4px 2px 1px 5px;
    + Margin:4px 2px 1px 5px;      

Document flow
* there are two types of elements: Block and Inline
* a block element 
  + occupy all the horizontal space of its parent element and the vertical hight of its content
  + has a new line before and after itself
    + if you have a block element in the middle of two inline elements, then you will have each element on its own line
  + multiple block level elements will stack on top of each other
  + examples include headings, div and form
* inline element
  + occupy only the width and height of their content
  + they don't appear on a new line, hence the name inline
  + multiple inline elements can form a row of elements
  + examples include anchor, img, input, label, b(bold), i(italic), em and span
* it is possible to change elements from inline to block and vice versa using display property
  + for example, you can change the display property of a div element with id of middle by
```css
#middle {
    display: inline;
}
```

Introduction to responsive design
* responsive design means a web page can automatically stretch or shrink depending on the screen it is desplayed on
* screen resolution is expressed by number of pixels in width times number of pixels in height (1920 times 1080)
* responsive design is a set of 3 practices that allows a website to automatically change its visuals to respond based on the device it is displayed on
  + flexible grids
    + flexible grids are made up of columns, gutters (space between columns) and margins (space between content and the left and right edges of screen are margins)
    + instead of defining website element sizes based on pixels, flexible grids are defined in percentage values, allowing them to adjust depending on screen size 
  + fluid images
    + set the max width property of images to 100%
    + images will scale down smaller if their containing columns becomes narrower than the images size but never grow larger
      + this enables the images to scale down to fit in a flexible sized column rather than overflow it 
  + media queries
    + as a port of CSS
    + allows to query the display size, orientation and aspect ratio to conditionally apply CSS rules
    + the following example will set the screen color to blue if the screen size <= 700 px
    
    ```CSS
@media only screen and (max-width: 700px) {
    body {
        background-color: blue;
    }
}
```

* combining flexible grids, fluid images and media queries delivers responsive grid
* Breakpoint
  + a breakpoint is the point at which a website's content and layout will adapt to provide the best possible user experience
  + a breakpoint can function in different ways across three different grids. Theser grids are
    + a fixed grid
      + fixed width columns and flexible margins
      + has a fixed content width that doesn't change in a specific breakpoint range while the flexible margins occupy the remaining space on sreeen
    + fluid or full width grid
      + fluid column widths
      + fixed gutters and side margins
      + has fluid content width that goes edge to edge as per the screen size
      + columns either grow or shrink to adapt to the available space
    + hybrid grid
      + have both fluid and fixed grid components
* many responsive design framework provide multiple CSS rules based on different device sizes to provide the best visual experience

#### Introduction to React
* Static and dynamic content
  + web server send HTML document to web browser
  + the content web server sends can be static and dynamic
  + static contents are files the server transfers just as they are stored on the web server such as videos and images
  + dynamic contents are generated when the HTTP request is made
    + it may generated based on input from a user
    + or based on the current date
    + web server communicates with application server/backend, which generates the dynamic content that web server sends back to user's browser
* how web site updates static and dynamic content
  + static content
    + web server send the files to browser
  + dynamic content
    + contents are generated by application server, which usually takes a longer time
    + application server performs more complex processing than web servers such as
      + communicate with database
      + run application logic
      + check permissions
    + there are application servers that only process specific tasks to imporve performance
    + application servers usually have a limited capacity of how many requests they can process per second
    + web servers use a process called caching as a saved copy of dynamic content
      + reduce the workload of application servers
      + reduce the latency sinde the contents are readily available upon request      

* Single page applications (SPA)
  + traditional web sites were built on multiple web pages, which are
    + resource intensive to web server 
      + sending entire web pages for every request consumes excessive bandwidth
      + generating dynamic pages consumes CPU time
  + in SPA
    + only one HTML page that is sent from web server to the browser
    + this page has multiple updates to its content as user interacting with the web site
    + A SPA allows the user to interact with the web site without the application needing to download entire new web pages
    + it rewirtes the current web page as the user interacts with it
    + browsing experience that feels faster and more responsive to user input
* SPA has two approaches to serving code and resources
  + bundling
    + when browser requests the application, the server returns and loads all necessary HTML, CSS and JS immediately
  + lazy loading or code splitting
    + when browser requests the application, the server only returns minimum HTML, CSS and JS to load the application
      + additional resources are downloaded as required
  + when a user navigates a specific section of the application, both approaches are valid. the choice depends on size, complexity and bandwidth requirements of the application
    + if app is complicated and requires a lot of resources, the bundle will be large and take a long time to download
    + if only a section needs to be update, SPA can send a json object to update only that section, which is more efficient and faster
    + if you need to update the majority of the page to display a different form of content, you can design a template and only updates the contents that are updated by SPA    

* What is React?
  + React is focused on working with components which makes it simple to build functional user interfaces on web and mobile
  + React is an open source library with core contributors and companies that maintain it 
  + Developers use React to develop SPAs and you can also develop mobile apps with React native
  + React is used for user interface and used in conjunction with other JS libraries during developments
  + allows developers to use less code to develop functionality
* react components
  + a small piece of a user interface, such as a music player or photo gallery. 
  + components allow you to do
    + isolated development
    + isolated testing
    + re-using components
  + there are many component libraries available for you to use in your apps  

* How react works?
  + react builds out its tree of components
  + it builds its tree of components, it builds out its own DOM in memory, called the virtual DOM
  + virtual DOM is the representation of browser DOM that is kept in memory
  + react uses this virtual DOM to update the browser DOM, only when it needs to. This ensures your app is fast and responsive to user inputs
  + React checks if HTML component in the virtual DOM matches the browser DOM, if a change is required, the browser DOM is updated, otherwise, no updates are performed. This is call reconciliation 
* when you upadate a component in React:
  + the virtual DOM is updated
  + React then compares the virtual DOM to the previous version of virtual DOM, and determines which components have been changed
  + if a change occurs, only that component is updated in the browser DOM
  + changes on the browser DOM cause the displayed webpage to change
* as a result, only changed DOM components are updated, instead of changing the entire web page

* Component hierarchy
  + every react application contains at least one component, the root component, or app component
  + components are added to app component to build out the tree structure to make up the application
  + app component is the entire web page
  + app component has two child components
    + navbar component that contains the title of the blog, navigation links for the website along with the search component
    + page component contains main feature component, and two instances of a small feature components
      + the page actually contains 4 small feature components, 2 of them are reused, but with different properties
